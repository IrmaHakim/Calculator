```python
import tkinter as tk
from tkinter import*


Calc=Tk()
Calc.title("Calculator")
operator=""
cal_input=StringVar()

def btn_click(numbers):
    global operator
    operator=operator + str(numbers)
    cal_input.set(operator)


def btn_clear():
    global operator
    operator=""
    cal_input.set("")
    
    
def btn_equal():
    global operator
    ans=str(eval(operator))
    cal_input.set(ans)

screen=Entry(Calc,textvariable=cal_input,fg='black',bg='powder blue',font='arial 18 bold ',width='12',bd="10",justify="right",insertwidth=5)
screen.grid(row=0,columnspan=5)


a=Button(Calc,text="7",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(7))
a.grid(row=2,column=0)
b=Button(Calc,text="8",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(8))
b.grid(row=2,column=1)
c=Button(Calc,text="9",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(9))
c.grid(row=2,column=2)
d=Button(Calc,text="+",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click("+"))
d.grid(row=2,column=3)
e=Button(Calc,text="4",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(4))
e.grid(row=3,column=0)
f=Button(Calc,text="5",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(5))
f.grid(row=3,column=1)
g=Button(Calc,text="6",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(6))
g.grid(row=3,column=2)
h=Button(Calc,text="-",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click("-"))
h.grid(row=3,column=3)
i=Button(Calc,text="1",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(1))
i.grid(row=4,column=0)
j=Button(Calc,text="2",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(2))
j.grid(row=4,column=1)
k=Button(Calc,text="3",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(3))
k.grid(row=4,column=2)
l=Button(Calc,text="*",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click("*"))
l.grid(row=4,column=3)
m=Button(Calc,text="C",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= btn_clear)
m.grid(row=5,column=0)
n=Button(Calc,text="0",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click(0))
n.grid(row=5,column=1)
o=Button(Calc,text="/",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= lambda: btn_click("/"))
o.grid(row=5,column=2)
p=Button(Calc,text="=",padx=12,bg="white",fg="black",font='arial 18 bold  ',bd=8,command= btn_equal)
p.grid(row=5,column=3)


Calc.mainloop()
```
